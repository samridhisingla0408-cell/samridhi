```python
import numpy as np
```

```python
from collections import Counter
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
x_train=np.array([[1],[2],[3],[6],[7],[8]])
```

```python
y_train=np.array([0,0,0,1,1,1])
```

```python
test=np.array([5])
```

```python
distance = []
for x in x_train :
    d= np.sqrt(np.sum(x-test)**2)
    distance.append(d)
print(distance)
```

```python
k=3
```

```python
nearest = np.argsort(distance)[:k]
labels = y_train[nearest]
prediction = Counter(labels).most_common(1)[0][0]
print("Nearest neighbours:",nearest)
print("labels:",labels)
print("prediction:",prediction)
```

```python
plt.scatter(x_train, y_train)
plt.scatter (test , prediction , color='red')
plt.xlabel("x values")
plt.ylabel("class")
plt.title("KNN Classification")
plt.show()
```

```python
plt.figure(figsize=(8,5))
#plot class 0
plt.scatter (x_train[y_train==0],np.zeros(3),color='blue',s=120,label='Class0')
#plot class 1
plt.scatter (x_train[y_train==1],np.zeros(3),color='green',s=120,label='Class1')
#plot test point
plt.scatter (test,[0.5],color='red',s=180, marker='*',label='Test Point')
plt.title("Training Data and Test Point")
plt.xlabel("Feature Value")
plt.ylabel("Class")
plt.yticks([0,1],["Class 0","Class 1"])
plt.grid(True)
plt.legend()
plt.show()
```
