```python
x=np.array([1,2,3,4,5])
```

```python
y=np.array([2,3,4,5,6])
```

```python
mean_x= np.mean(x)
print(mean_x)
```

```python
mean_y = np.mean(y)
print(mean_y)
```

```python
num = np.sum((x-mean_x)*(y-mean_y))
print(num)
```

```python
den=np.sum((x-mean_x)**2)
print(den)
```

```python
m=num/den
print("slope",m)
```

```python
b=mean_y-m*mean_x
print("intreupt",b)
```

```python
y_pred=m*x +b
```

```python
plt.scatter(x,y,color="red")
```

```python
plt.plot(x,y_pred,color="black")
```

```python
plt.scatter(x,y,color="red")
plt.plot(x,y_pred,color="black")
```
